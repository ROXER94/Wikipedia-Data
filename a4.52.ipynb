```python
#http://docs.python.org/2/library/time.html
from time import time,strptime,strftime,gmtime
from calendar import timegm

def ts2utc(ts):    #convert timestring to utc
    return int(timegm(strptime(ts,'%H:%M, %d %B %Y')))

def utc2ts(utc):   #convert utc to timestring
    return strftime('%d/%b/%Y:%H:%M',gmtime(utc))

min=60
hr=60*60
day=24*hr

sep30=ts2utc('00:00, 30 September 2014')  # = 1412035200
```

```python
figure(figsize=(8,6))
hist([ts2utc(ts) for ts in wikidata],bins=range(sep30,sep30+3*day+hr,hr))
xt=range(sep30,sep30+3*day,6*hr)
xticks(xt,[utc2ts(utc) for utc in xt],rotation=300,ha='left')
xlim(sep30,sep30+3*day)
ylabel('# edits / hour');
```

```python
from collections import Counter
from scipy.misc import factorial
def poisson(m,z=1): return exp(-z)*z**m/factorial(m)
```

```python
def myHist(data,hours):
  mincounts=Counter(data).values()
  mmax=max(mincounts)
  totalmin=len(set([ts[:2] for ts in data]))*60*3
  avg = len(data)/float(totalmin)
  zeroes = totalmin - len(mincounts)

  hist(mincounts + [0]*zeroes, arange(-.5,mmax+1), label='data')
  plot(totalmin*poisson(arange(mmax+2),avg),'ro-',linewidth=1.5,
       label='Poisson, avg='+str(round(avg,2))+'/min')
  xlim(-.5,mmax+1)
  xlabel('#edits / minute')
  ylabel('#minutes of '+str(totalmin)+' sampled')
  title("Wikipedia editing contributions for "+hours+' hours, 30 Sep - 2 Oct 2014')
  legend()
```

```python
figure(figsize=(18,12))
subplot(221)
myHist(low,'night')
```

```python
figure(figsize=(18,12))
subplot(222)
myHist(high,'day')
```

```python
figure(figsize=(18,12))
subplot(223)
myHist(intermed1,'evening')
```

```python
figure(figsize=(18,12))
subplot(224)
myHist(intermed2,'early am')
```
